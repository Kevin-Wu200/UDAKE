# 用户验证与模型自评估系统教程

本教程演示阶段11新增模块的最小闭环：实时评估 -> 模型选择 -> 优化触发 -> 报告生成。

```python
from services.backend.app.services.self_evaluation_service import SelfEvaluationService
service = SelfEvaluationService()
```

```python
records = []
for i in range(8):
    pred = 12.0 + i * 0.2
    actual = pred + (0.1 if i % 2 == 0 else -0.12)
    records.append({
        'evaluation_id': f'nb_{i}',
        'dataset_id': 'notebook_demo',
        'model_id': 'st_transformer',
        'module': 'realtime_interpolation',
        'x': 120 + i * 0.02,
        'y': 30 + i * 0.02,
        'predicted_value': pred,
        'actual_value': actual,
        'result': 'accept',
        'confidence': 0.65,
        'uncertainty': 0.15
    })
result = service.evaluate_realtime({'records': records})
result['performance']['overall_accuracy']
```

```python
selection = service.select_best_model({
    'candidates': [
        {'model_id': 'm1', 'performance_score': 0.72, 'uncertainty_score': 0.24, 'scenario_score': 0.60},
        {'model_id': 'm2', 'performance_score': 0.82, 'uncertainty_score': 0.18, 'scenario_score': 0.68}
    ]
})
selection['selected_model']
```

```python
task = service.trigger_optimization({'trigger_type': 'manual', 'async': False, 'data_volume': 120})
report = service.generate_report({'report_type': 'all', 'format': 'json'})
task['status'], report['report_id']
```
